# Implement Lower Bound

Problem Statement: Given a sorted array of N integers and an integer x, write a program to find the lower bound of x.

What is lower bound?
The lower bound algorithm finds the first or the smallest index in a sorted array where the value at that index is greater than or equal to a given key i.e. x.

The lower bound is the smallest index, ind, where arr[ind] >= x. But if any such index is not found, the lower bound algorithm returns n i.e. size of the given array.

```python
class LowerBound:
    @staticmethod
    def lower_bound_finder(nums: List[int], x: int) -> int:
        # brute force linear algo
        # n = len(nums)
        # for i in range(n):
        #     if nums[i] >= x:
        #         return i
        # return n
    
        # binary search algo
        start = 0
        end = len(nums) - 1
        ans = len(nums)
        while(start <= end):
            mid = (start + end) // 2
            if nums[mid] >= x:
                ans = mid
                end = mid - 1 
            else: 
                start = mid + 1
        return ans


sol = LowerBound()
array = [3,5,8,15,19]
print(sol.lower_bound_finder(array, 9))
```

# Implement Upper Bound

Problem Statement: Given a sorted array of N integers and an integer x, write a program to find the upper bound of x.

What is Upper Bound?
The upper bound algorithm finds the first or the smallest index in a sorted array where the value at that index is greater than the given key i.e. x.

The upper bound is the smallest index, ind, where arr[ind] > x.

```python
class UpperBound:
    @staticmethod
    def upper_bound(nums: List[int], x: int) -> int:
        ans = len(nums)
        start = 0
        end = len(nums) - 1
        while (start <= end):
            mid = (start + end) // 2
            if nums[mid] > x:
                end = mid -1
                ans = mid
            else: start = mid + 1

        return ans
    
sol = UpperBound()
array = [1, 2, 3, 5, 6, 7]
x = 4
print(sol.upper_bound(array, x))
```

# Floor and Ceil in Sorted Array

Problem Statement: You're given an sorted array arr of n integers and an integer x. Find the floor and ceiling of x in arr[0..n-1]. The floor of x is the largest element in the array which is smaller than or equal to x. The ceiling of x is the smallest element in the array greater than or equal to x

```python
class FloorandCeil:
    @staticmethod
    def find_floor_and_ceil(nums: List[int], x: int) -> Tuple[int, int]:
        # find floor
        start = 0
        end = len(nums) - 1
        floor_num, ceil_num = -1, -1 # as floor and ceil might not exist
        # floor calculation
        while(start <= end):
            mid = (start + end) // 2
            if nums[mid] <= x:
                floor_num = nums[mid]
                start = mid + 1
            else: end = mid - 1
        # for ceil calculation
        start = 0
        end = len(nums) - 1
        while(start <= end):
            mid = (start + end) // 2
            if nums[mid] >= x:
                ceil_num = nums[mid]
                end = mid - 1
            else: start = mid + 1


        ## exercise do it in one pass

        return floor_num, ceil_num
        

```
