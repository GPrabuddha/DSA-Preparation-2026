# Find Second Smallest and Second Largest Element in an array

Problem Statement: Given an array, find the second smallest and second largest element in the array. Print ‘-1’ in the event that either of them doesn’t exist.

```python
class Solution():
    def SecondSmallestSecondLargest(self, nums: List[int]) -> int | tuple[int, int]:
        # Brute force #

        # tip: check edge cases, here I didn't consider if len(array) = 1 or 0

        # time complexity: O(nlogn) and space complexity: O(1)
        # fails for repeated numbers
        # n = len(nums)
        # if n == 0 or n == 1:
        #     return -1, -1
        # nums.sort()
        # if nums[1] and nums[-2]:
        #     return nums[1], nums[-2]
        # else: return -1

        # Better complexity + accomodate repeated elements #

        # tip: always remember sort incurs O(nlogn) and iterating array O(n) therefore if iteration can solve then use that
        # Time complexity O(n) and space complexity O(1)

        n = len(nums)
        if n == 0 or n == 1:
            return -1, -1
        smallest_number = largest_number = nums[0]
        # calculate largest first
        for i in nums[1:]:
            if i > largest_number:
                largest_number = i
            if i < smallest_number:
                smallest_number = i
        # calculate 2nd largest
        second_largest = second_smallest = nums[0]
        for i in nums[1:]:
            if i > second_largest and i < largest_number:
                second_largest = i
            if i < second_smallest and i > smallest_number:
                second_smallest = i
        return second_smallest, second_largest



sol = Solution()
input1 = [1, 2, 4, 7, 7, 5]
ans = sol.SecondSmallestSecondLargest(input1)
print(ans)

```

# Check if an Array is Sorted

Problem Statement: Given an array of size n, write a program to check if the given array is sorted in (ascending / Increasing / Non-decreasing) order or not. If the array is sorted then return True, Else return False.

```python
class Solution():
    def ifarraysorted(self, nums: List[int]) -> bool:
        # Optimal #
        # time complexity O(n) and space complexity is O(1)
        for index, item in enumerate(nums[:-1]):
            if item > nums[index+1]:
                return False
        return True

sol = Solution()
# array = [1,2,3,4,5]
array = [5,4,6,7,8]
ans = sol.ifarraysorted(array)
print(ans)
```

# Remove Duplicates in-place from Sorted Array

Problem Statement: Given an integer array sorted in non-decreasing order, remove the duplicates in place such that each unique element appears only once. The relative order of the elements should be kept the same.

If there are k elements after removing the duplicates, then the first k elements of the array should hold the final result. It does not matter what you leave beyond the first k elements.

```python
class Solution():
    def RemoveDuplicates(self, nums: List[int]) -> List[int]:
        # first remove the duplicates, takes O(n) time
        for index, item in enumerate(nums[:-1]):
            if item == nums[index+1]:
                nums[index] = None

        # rearrange array, should be done in O(n) time

        # tip: use 2nd pointer if we want to keep track of certain states or elements
        # here I could have created a pointer to track unique element to avoid looking at solution

        for index, item in enumerate(nums):
            unique_element = 0  # let first element be unique
            if item == nums[index + 1]:
                nums[index] = None
            unique_element = index + 1


        
        return nums
    
sol = Solution()
# array = [1,1,2,2,2,3,3]
array = [1,1,1,2,2,3,3,3,3,4,4]
ans = sol.RemoveDuplicates(array)
print(ans)
```

# Left Rotate the Array by One

Problem Statement: Given an integer array nums, rotate the array to the left by one.

Note: There is no need to return anything, just modify the given array.

```python
class Solution():
    def rotate_by_one(self, nums: List[int]) -> None:

        # time complexity O(n) and space complexity O(1)
        for index in range(len(nums)-1):
            temp = nums[index]
            nums[index] = nums[index + 1] 
            nums[index + 1] = temp
        return nums
    
sol = Solution()
array = [1, 2, 3, 4, 5] 
ans = sol.rotate_by_one(array)
print(ans)
```

# Rotate array by K elements
Problem Statement: Given an array of integers, rotating array of elements by k elements either left or right.

```python
class Solution():
    def rotate_by_k(self, nums: List[int], k: int, RorL: str)-> None:
        # let's start with rotate left by k
        # Brute: we can do rotate by 1; k times, will take O(nk)
        # I just read that modulus % can be used for cyclic structures like clock, array etc so let's try that for better complexity

        # tip: for array questions don't shy from using temporary arrays, you can optimise space later so don't shy from using

        # below soln take O(n) time complexity but O(n) space complexity as we create a temporary array of size n
        n = len(nums)
        temp_array = []
        if RorL.lower() == "left":
            for i in range(k, n):
                temp_array.append(nums[i])
            for i in range(k):
                temp_array.append(nums[i])

        elif RorL.lower() == "right":
            for i in range(n - k, n):
                temp_array.append(nums[i])
            for i in range(n-k):
                temp_array.append(nums[i])
        return temp_array
    
sol = Solution()
array = [1, 2, 3, 4, 5, 6, 7]
ans = sol.rotate_by_k(array, 2, "right")
print(ans)
    
```

# Count Maximum Consecutive One's in the array
Problem Statement: Given an array that contains only 1 and 0 return the count of maximum consecutive ones in the array..



```python
class Solution():
    def MaxConsecutiveOne(self, nums: List[int]) -> int:
        current_max_sum = 1
        max_sum = 0
        for i in range(len(nums)-1):
            if nums[i] == 1 and nums[i + 1] == 1:
                current_max_sum += 1
            elif nums[i] == 0:
                if current_max_sum > max_sum:
                    max_sum = current_max_sum
                current_max_sum = 1
        return max_sum
    
sol = Solution()
array = [1, 1, 0, 1, 1, 1]
ans = sol.MaxConsecutiveOne(array)
print(ans)
```
