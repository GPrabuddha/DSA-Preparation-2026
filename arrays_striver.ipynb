# Largest Element

Problem Statement: Given an array, we have to find the largest element in the array.

```python
class Solution():
    def LargestElement(self, nums: List[int]) -> int:
        # optimal #

        # O(n) time complexity, O(1) space complexity

        ## largest_number = 0 # this runs into issue if array only contains negative elements
        largest_number = nums[0]
        for i in nums[1:]:
            if i > largest_number:
                largest_number = i
        return largest_number

        # Brute force #
        # tip: use sort function
        # O(nlogn) time complexity, O(1) space complexity

        # nums.sort()
        # return(nums[-1])




sol = Solution()
ans = sol.LargestElement([1, 2, 5, 8, -5])
print(ans)
```

# Find Second Smallest and Second Largest Element in an array

Problem Statement: Given an array, find the second smallest and second largest element in the array. Print ‘-1’ in the event that either of them doesn’t exist.

```python
class Solution():
    def SecondSmallestSecondLargest(self, nums: List[int]) -> int | tuple[int, int]:
        # Brute force #

        # tip: check edge cases, here I didn't consider if len(array) = 1 or 0

        # time complexity: O(nlogn) and space complexity: O(1)
        # fails for repeated numbers
        # n = len(nums)
        # if n == 0 or n == 1:
        #     return -1, -1
        # nums.sort()
        # if nums[1] and nums[-2]:
        #     return nums[1], nums[-2]
        # else: return -1

        # Better complexity + accomodate repeated elements #

        # tip: always remember sort incurs O(nlogn) and iterating array O(n) therefore if iteration can solve then use that
        # Time complexity O(n) and space complexity O(1)

        n = len(nums)
        if n == 0 or n == 1:
            return -1, -1
        smallest_number = largest_number = nums[0]
        # calculate largest first
        for i in nums[1:]:
            if i > largest_number:
                largest_number = i
            if i < smallest_number:
                smallest_number = i
        # calculate 2nd largest
        second_largest = second_smallest = nums[0]
        for i in nums[1:]:
            if i > second_largest and i < largest_number:
                second_largest = i
            if i < second_smallest and i > smallest_number:
                second_smallest = i
        return second_smallest, second_largest



sol = Solution()
input1 = [1, 2, 4, 7, 7, 5]
ans = sol.SecondSmallestSecondLargest(input1)
print(ans)

```

# Check if an Array is Sorted

Problem Statement: Given an array of size n, write a program to check if the given array is sorted in (ascending / Increasing / Non-decreasing) order or not. If the array is sorted then return True, Else return False.

```python
class Solution():
    def ifarraysorted(self, nums: List[int]) -> bool:
        # Optimal #
        # time complexity O(n) and space complexity is O(1)
        for index, item in enumerate(nums[:-1]):
            if item > nums[index+1]:
                return False
        return True

sol = Solution()
# array = [1,2,3,4,5]
array = [5,4,6,7,8]
ans = sol.ifarraysorted(array)
print(ans)
```

# Remove Duplicates in-place from Sorted Array

Problem Statement: Given an integer array sorted in non-decreasing order, remove the duplicates in place such that each unique element appears only once. The relative order of the elements should be kept the same.
